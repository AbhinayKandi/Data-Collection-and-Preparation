## Visualization

```python
import matplotlib.pyplot as plt
import matplotlib.pylab as pyl
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
%pylab inline
```

```python
#historical view of the closing price for Apple
apple['apple_high'].plot(fontsize=15,title='Time Series High-price of Apple',legend=True,grid = True,figsize=(10,4),color='green')
plt.xlabel('Date')
plt.ylabel('High-Price')
```

I see that lowest high-price was recorded at around 156 in the month of February 2018 and the highest high-price was recorded at around 185 in the month of March 2018 for the latest 100 historical days.

```python
colors=['r','b']
apple.plot(legend=True,kind='scatter',x='apple_close',y='apple_adjusted_close',c=colors,title='Apple Close and Adjusted Close',figsize=(10,4))
plt.legend=True
plt.show()
```

There seems to be negligible difference in market close price and adjusted(company) close price. 

```python
apple['apple_volume'].plot(fontsize=15,figsize=(10,4),legend=True,title="Stock Volume of Apple",style="-.",color='red')
plt.ylabel("Volume")
```

The highest Apple stock was traded in the month of February 2018.

```python
#Moving average for 20 days and 50 days Moving average of Apple stock
#Moving Average is calculated by taking the average closing price of a stock over the last "x" periods.
apple['MA20'] = pd.rolling_mean(apple['apple_adjusted_close'],20)
apple['MA50'] = pd.rolling_mean(apple['apple_adjusted_close'], 50)
apple[['apple_adjusted_close', 'MA20', 'MA50']].plot(fontsize=15,title='Moving average for 20 days and 50 days Moving average of Apple stock',grid=True,subplots=False, figsize=(10, 4))
plt.ylabel("Adjusted_close_price")
```

Looking at the 2 Moving Averages(Simple), the direction of 20-Day MA(Orange line) is angled up and the direction of 50-Day MA(Green line) is moving slightly sideways. By this, trader would like to buy Apple stock. But looking at the overall trend(Orange line), the trader would not buy as the price is angled down. MA's would help in the short-term trading.

```python
# Apple's historical percentage daily returns
apple_return=apple['apple_adjusted_close'].pct_change()
apple_return.plot(fontsize=15,legend=True,title="Daily Return of Apple",figsize=(10,4),linestyle='--',marker='o').axhline(y = 0, color = "black", lw = 2)
plt.ylabel("Adjusted_close_price")
```

I see that the returns of Apple stock are not so great and show a downward trend from mid January 2018. I'm more concerned about the relative change of an asset rather than its absolute price. 

```python
# Histogram to visualise the percentage daily returns of Apple
sns.distplot(apple_return.dropna(),bins=100,color='purple')
plt.xlabel("Daily Return")
plt.title("Histogram of relative percentage daily returns of Apple")
```

```python
apple_return.dropna().quantile(0.05)
```

The 0.05 empirical quantile of Apple stock daily returns is at -0.022. That means that with 95% confidence, our worst daily loss will not exceed 2.2%. If we have a 1 million Euros investment, our one-day 5% VaR is 0.022 * 1,000,000 = 22,000 Euros.

## Final Manipulation, Exploration and Visualization

I have subsetted the original dataframes and concatenated into single Dataframe with 3 columns each of 3 stocks.

```python
apple_adjclose=apple[['apple_adjusted_close']]
fb_adjclose=fb[['fb_adjusted_close']]
mst_adjclose= mst[['mst_adjusted_close']]
stock=pd.concat([apple_adjclose,fb_adjclose,mst_adjclose],axis=1) # This dataframe would be used for Visualisations.
stock.tail()
```

```python
#historical adjusted close price of apple, facebook and microsoft stock.
stock.plot(fontsize=15,figsize=(10,4),title="Adjusted close price of Apple, Facebook and Microsoft")
plt.ylabel("Adjusted_close_price")
```

While absolute price is important; pricy stocks are difficult to purchase, which affects not only their volatility but our ability to trade that stock. Apple's and Favcebook's stocks are much more expensive than Microsft's, and this difference makes Micorosoft's stocks appear much less than they are.

```python
#historical adjusted close price of apple, facebook and microsoft stock.
stock.plot(fontsize=13,figsize=(10,4),secondary_y=["apple_adjusted_close","fb_adjusted_close"],grid=True,title="Adjusted close price of Apple, Facebook and Microsoft")
plt.ylabel("Adjusted_close_price")
```

The solution to the previous chart is to use two different scales when plotting the data; one scale will be used by Apple and Facebook stocks, and the other by Microsoft. From the above chart, I see that Microsoft has better adjusted close price(compared to that on 2017-10-30) among the other stocks which was obscure in the previous chart.

```python
#to find the percent change for each day
stock_return=stock.pct_change()
stock_return.plot(fontsize=15,figsize=(10,4),grid = True).axhline(y = 0, color = "black", lw = 2) # axhline is the horizontal line
plt.ylabel("Adjusted_close_price")
```

If I had bought shares these 3 stocks on 2017-10-30; then I would be in loss today (Downward Trend) for facebook stock and profit for other two stocks.

```python
#Joinplot to compare the daily returns of Apple and Facebook
sns.jointplot('apple_adjusted_close','fb_adjusted_close',stock_return, kind='scatter',size=5, color='skyblue')
```

The pearsonr value also known as the Pearson product-moment correlation coefficient shows how correlated the daily percentage returns are. There is 46% correlation between Apple and Facebook Stock on adjusted close price which is not a good relation as I predicted.

```python
# Correlation analysis for every possible combination of stocks
sns.pairplot(stock_return.dropna(),size=3)
```

The above plot shows correlation between every stock and its histogram in my Dataframe.

```python
#Value at risk using the "Bootstrap" method
stock_return.dropna().quantile(0.05)
```

The 0.05 empirical quantile of Apple stock daily returns is at -0.022. That means that with 95% confidence, our worst daily loss will not exceed 2.2%. If we have a 1 million Euros investment, our one-day 5% VaR is 0.022 * 1,000,000 = 22,000 Euros.

The 0.05 empirical quantile of Facebook stock daily returns is at -0.040. That means that with 95% confidence, our worst daily loss will not exceed 4.0%. If we have a 1 million Euros investment, our one-day 5% VaR is 0.040 * 1,000,000 = 40,000 Euros.

The 0.05 empirical quantile of Microsoft stock daily returns is at -0.026. That means that with 95% confidence, our worst daily loss will not exceed 2.6%. If we have a 1 million Euros investment, our one-day 5% VaR is 0.026 * 1,000,000 = 26,000 Euros.
